```python
import matplotlib.pylab as plt
# import networkx as nx
import numpy as np
import pandas as pd
import time
import sys
import os
import textwrap
import re
```

```python
filepath = "../Parliament Australia/every_politician_database-popolo-v1.0.json"

# source: http://everypolitician.org/uk/commons/download.html
#filepath = "../Parliament UK/everypolitician_UK-popolo-v1.0.json"

#pd.read_json(filepath)
import json

# Reading the json as a dict
with open(filepath) as json_data:
    data = json.load(json_data)
```

```python
members_of_parliament = [d['person_id'] for d in data['memberships']
                         if d['organization_id'] == '2ab5a3aa-ef1e-46c0-abc9-dad161aa0390' and
                            d['legislative_period_id'] == 'term/45']
print(len(members_of_parliament))
```

```python
df = pd.DataFrame(columns=["surname", "first_name", "name", "birth_date", "twitter_url", "handle",
                           "wikipedia", "in_current_parliament"])

for i, person in enumerate(data['persons']):
    print(person)
    twitter = None
    twitter_handle = None
    wikipedia = None
    if person['id'] in members_of_parliament:
        in_parliament = 1
    else:
        in_parliament = 0
    
    for link in person['links']:
        if link['note'] == 'twitter':
            twitter = link['url']
            twitter_handle = "@" + twitter.split('/')[3]
        elif link['note'] == 'wikipedia':
            wikipedia = link['url']
    df.loc[i] = [person['family_name'], person['given_name'], person['name'],
                 person['birth_date'], twitter, twitter_handle, wikipedia, in_parliament]
```

```python
print(df['twitter_url'].count())
print(df['in_current_parliament'].sum())
```

```python
counter = 0
cp_df = pd.DataFrame(columns=["surname", "first_name", "name", "birth_date", "twitter_url", "handle",
                           "wikipedia", "in_current_parliament"])

for i, row in df.iterrows():
    if row['handle'] and row['in_current_parliament']:
        print(row['handle'])
        cp_df.loc[i] = row
        counter += 1
print(counter)
cp_df = cp_df.reset_index()
```

```python
cp_df.to_csv("Twitter_list_Australian_MPs_everypolitician.csv")
```
